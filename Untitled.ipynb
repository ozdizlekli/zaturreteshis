```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16, ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import os
import cv2

# TensorFlow sürümünü ve GPU durumunu kontrol et
print(f"TensorFlow: {tf.__version__}")
print(f"GPU: {tf.config.list_physical_devices('GPU')}")

# Ana veri yolları
# Veri setinizin ana klasörüne işaret eder.
train_dir = 'chest_xray/train'
test_dir = 'chest_xray/test'
val_dir = 'chest_xray/val'
```

```python
# Tüm görüntülerin yeniden boyutlandırılacağı boyut
IMG_SIZE = 224
# Modele aynı anda verilecek görüntü sayısı
BATCH_SIZE = 32

# Veri artırma (Data Augmentation) için veri üreteci
# Bu işlem, modelin aşırı öğrenmesini engellemek için eğitim verisine rastgele dönüşümler uygular.
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Piksel değerlerini 0-1 arasına normalize etme
    rotation_range=20,       # Görüntüleri rastgele 20 derece döndürme
    width_shift_range=0.2,   # Görüntüleri yatayda kaydırma
    height_shift_range=0.2,  # Görüntüleri dikeyde kaydırma
    shear_range=0.2,         # Görüntüleri kesme
    zoom_range=0.2,          # Görüntülere yakınlaşma/uzaklaşma
    horizontal_flip=True,    # Görüntüleri yatayda çevirme
    fill_mode='nearest'      # Dönüşüm sonrası oluşan boşlukları doldurma
)

# Doğrulama ve test verisi için sadece normalizasyon yapılır, augmentation yapılmaz.
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Veri üreteçlerini oluştur
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',   # 2 sınıf (NORMAL ve PNEUMONIA) olduğu için binary
    shuffle=True           # Eğitim verisini karıştır
)

val_generator = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False          # Doğrulama verisi karıştırılmaz
)

test_generator = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False          # Test verisi karıştırılmaz
)

print(f"✅ Eğitim: {train_generator.samples} örnek")
print(f"✅ Doğrulama: {val_generator.samples} örnek")
print(f"✅ Test: {test_generator.samples} örnek")
```

```python
def create_transfer_learning_model(base_model_name='VGG16', input_shape=(224, 224, 3)):
    """Aktarım öğrenimi için temel model mimarisini oluşturur."""
    if base_model_name == 'VGG16':
        base_model = VGG16(
            weights='imagenet',       # ImageNet veri setinde önceden eğitilmiş ağırlıkları yükle
            include_top=False,        # Sınıflandırma katmanlarını çıkarma
            input_shape=input_shape   # Görüntü boyutunu belirleme
        )
    elif base_model_name == 'ResNet50':
        base_model = ResNet50(
            weights='imagenet',
            include_top=False,
            input_shape=input_shape
        )
    
    # Temel modelin katmanlarını dondurma
    # Bu, ilk eğitim aşamasında modelin önceden öğrendiği özellikleri korur.
    base_model.trainable = False
    
    # Yeni sınıflandırma katmanlarını ekle
    x = base_model.output
    x = GlobalAveragePooling2D()(x) # Özellik haritasını düzleştirme
    x = Dense(512, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    # Son katman: 1 çıktı ve sigmoid aktivasyonu ile 0-1 arası olasılık tahmini
    predictions = Dense(1, activation='sigmoid')(x)
    
    # Yeni modeli oluştur
    model = Model(inputs=base_model.input, outputs=predictions)
    
    return model, base_model

# VGG16 modeli oluştur
model_vgg16, base_vgg16 = create_transfer_learning_model('VGG16')

# Model derleme
# Optimizer: En çok kullanılan Adam optimizer
# Loss: İkili sınıflandırma için Binary Crossentropy
model_vgg16.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

# Geri çağrı fonksiyonları (Callbacks)
# Eğitimi otomatik olarak yönetir.
callbacks_tl = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1),
    ModelCheckpoint('best_vgg16_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)
]

print("✅ Model oluşturuldu!")
```

```python
# AŞAMA 1: Sadece yeni katmanları eğit
# Modelin sonuna eklenen yeni katmanlar, öğrenmeye sıfırdan başlar.
print("\n" + "="*50)
print("AŞAMA 1: YENİ KATMANLARI EĞİTME")
print("="*50 + "\n")

EPOCHS_STAGE1 = 20
history_stage1 = model_vgg16.fit(
    train_generator,
    epochs=EPOCHS_STAGE1,
    validation_data=val_generator,
    callbacks=callbacks_tl,
    verbose=1
)

# AŞAMA 2: Fine-tuning
# Temel modelin son katmanlarını eğitime dahil ederek daha ince ayar yapılır.
print("\n" + "="*50)
print("AŞAMA 2: FINE-TUNING")
print("="*50 + "\n")

# Temel modelin son 15 katmanını eğitilebilir hale getir
base_vgg16.trainable = True
for layer in base_vgg16.layers[:15]:
    layer.trainable = False

# Daha düşük bir öğrenme oranıyla yeniden derle
model_vgg16.compile(
    optimizer=Adam(learning_rate=0.00001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

EPOCHS_STAGE2 = 30
history_stage2 = model_vgg16.fit(
    train_generator,
    epochs=EPOCHS_STAGE2,
    validation_data=val_generator,
    callbacks=callbacks_tl,
    verbose=1
)

print("\n🎉 Model eğitimi tamamlandı!")
```

```python
def evaluate_model_and_plot(model, test_generator, model_name='Model'):
    """Modeli test setinde değerlendirir ve metrikleri görselleştirir."""
    print("\n" + "="*50)
    print(f"{model_name} TEST SETİ DEĞERLENDİRMESİ")
    print("="*50 + "\n")
    
    test_generator.reset()
    # Test setindeki tahminleri al
    y_pred_proba = model.predict(test_generator, verbose=1)
    y_pred = (y_pred_proba > 0.5).astype(int).flatten()
    y_true = test_generator.classes
    
    target_names = ['NORMAL', 'PNEUMONIA']
    print("\nSınıflandırma Raporu:")
    print("-" * 50)
    print(classification_report(y_true, y_pred, target_names=target_names))
    
    # Confusion Matrix ve diğer metrikleri hesapla
    cm = confusion_matrix(y_true, y_pred)
    test_scores = model.evaluate(test_generator, verbose=0)
    
    print("\nTest Seti Metrikleri:")
    print("-" * 50)
    print(f"Test Accuracy: {test_scores[1]:.4f}")
    print(f"Test Precision: {test_scores[2]:.4f}")
    print(f"Test Recall: {test_scores[3]:.4f}")
    print(f"Test F1-Score: {2 * (test_scores[2] * test_scores[3]) / (test_scores[2] + test_scores[3]):.4f}")
    
    # Görselleştirmeler
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Confusion Matrix grafiği
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
                xticklabels=target_names, yticklabels=target_names)
    axes[0].set_title(f'{model_name} - Confusion Matrix', fontsize=14)
    axes[0].set_ylabel('Gerçek Değer')
    axes[0].set_xlabel('Tahmin')
    
    # ROC Eğrisi
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    axes[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Eğrisi (AUC = {roc_auc:.2f})')
    axes[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    axes[1].set_title(f'{model_name} - ROC Curve', fontsize=14)
    axes[1].set_xlabel('False Positive Rate')
    axes[1].set_ylabel('True Positive Rate')
    axes[1].legend(loc="lower right")
    plt.tight_layout()
    plt.show()

# Değerlendirme fonksiyonunu çağır
evaluate_model_and_plot(model_vgg16, test_generator, 'VGG16_Transfer_Learning')
```

```python
def find_last_conv_layer(model):
    """Modeldeki son konvolüsyon katmanını otomatik olarak bulur."""
    for layer in reversed(model.layers):
        if 'conv' in layer.name:
            return layer.name
    return None

def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    """Grad-CAM için ısı haritası oluşturan ana fonksiyon."""
    grad_model = Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def apply_gradcam_overlay(img_path, heatmap):
    """Isı haritasını orijinal görüntü üzerine bindirir."""
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(img, 0.6, heatmap, 0.4, 0)
    return cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB)

print("\n" + "="*50)
print("GRAD-CAM ANALİZİ BAŞLATIYOR")
print("="*50 + "\n")

# Hata ve doğru tahmin indekslerini bulma
test_generator.reset()
y_pred_proba = model_vgg16.predict(test_generator, verbose=0)
y_pred = (y_pred_proba > 0.5).astype(int).flatten()
y_true = test_generator.classes

false_pos_indices = np.where((y_pred == 1) & (y_true == 0))[0]
false_neg_indices = np.where((y_pred == 0) & (y_true == 1))[0]
correct_pos_indices = np.where((y_pred == 1) & (y_true == 1))[0]

last_conv_layer_name = find_last_conv_layer(model_vgg16)

# Rastgele örnek Grad-CAM görselleri seçme
sample_indices = np.concatenate((
    np.random.choice(false_pos_indices, min(2, len(false_pos_indices)), replace=False), # 2 Yanlış Pozitif
    np.random.choice(false_neg_indices, min(2, len(false_neg_indices)), replace=False), # 2 Yanlış Negatif
    np.random.choice(correct_pos_indices, min(2, len(correct_pos_indices)), replace=False) # 2 Doğru Pozitif
))

image_paths = [test_generator.filepaths[i] for i in sample_indices]

# Grad-CAM görselleştirmelerini çizme
plt.figure(figsize=(15, len(image_paths) * 5))
for i, img_path in enumerate(image_paths):
    img_array = image.img_to_array(image.load_img(img_path, target_size=(224, 224)))
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    heatmap = make_gradcam_heatmap(img_array, model_vgg16, last_conv_layer_name)
    superimposed_img = apply_gradcam_overlay(img_path, heatmap)

    plt.subplot(len(image_paths), 3, i*3 + 1)
    plt.imshow(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB))
    plt.title("Orijinal")
    plt.axis('off')

    plt.subplot(len(image_paths), 3, i*3 + 2)
    plt.imshow(heatmap, cmap='jet')
    plt.title("Isı Haritası")
    plt.axis('off')

    plt.subplot(len(image_paths), 3, i*3 + 3)
    plt.imshow(superimposed_img)
    plt.title("Grad-CAM Overlay")
    plt.axis('off')

plt.tight_layout()
plt.show()

print("\n✅ Grad-CAM görselleştirmesi tamamlandı!")
```

```python
import gradio as gr
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# ----- Adım 1: Modeli Yükle -----
# Eğitimden sonra kaydettiğin en iyi modeli yüklüyoruz
model_path = 'best_vgg16_model.h5'
try:
    best_model = load_model(model_path)
    print(f"✅ Model '{model_path}' başarıyla yüklendi!")
except Exception as e:
    print(f"Hata: Model yüklenemedi. Dosya yolunu kontrol edin. Hata: {e}")
    best_model = None

# ----- Adım 2: Tahmin Fonksiyonunu Oluştur -----
# Gradio arayüzünden gelen görüntüyü işleyip tahmin yapan ana fonksiyon
def predict_pneumonia(img):
    if best_model is None:
        return "Model yüklenemedi, lütfen bir yöneticinizle iletişime geçin."

    # Görüntüyü modele uygun formata getir
    img_resized = tf.image.resize(img, (224, 224))
    img_normalized = img_resized / 255.0  # Normalizasyon
    img_array = np.expand_dims(img_normalized, axis=0) # Batch boyutu ekle

    # Tahmin yap
    prediction = best_model.predict(img_array)[0][0]

    # Sonucu yorumla
    if prediction > 0.5:
        return f"Tahmin: Pnömoni | Olasılık: %{prediction*100:.2f}"
    else:
        return f"Tahmin: Normal | Olasılık: %{(1-prediction)*100:.2f}"

# ----- Adım 3: Gradio Arayüzünü Oluştur -----
# Arayüz için bir başlık ve açıklama
title = "Akciğer Röntgeni Pnömoni Tespiti Sistemi"
description = "Bir göğüs röntgeni görüntüsü yükleyin ve modelin pnömoni veya normal tahminini alın."

# Arayüz bileşenlerini tanımla
image_input = gr.Image(height=224, width=224)
label_output = gr.Label()

# Arayüzü başlat
if best_model is not None:
    gr.Interface(
        fn=predict_pneumonia,       # Hangi fonksiyonu kullanacağını belirt
        inputs=image_input,         # Girdi olarak resim alacağını belirt
        outputs=label_output,       # Çıktı olarak etiket döndüreceğini belirt
        title=title,                # Arayüz başlığı
        description=description,    # Açıklama metni
        live=False                  # Canlı güncellemeyi kapat (performans için)
    ).launch(share=True)
else:
    print("Model yüklenemediği için Gradio arayüzü başlatılamadı.")
```
